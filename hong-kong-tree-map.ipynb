```python
import folium
from folium import Circle, Marker
from folium.plugins import MarkerCluster

# Create a map
m_1 = folium.Map(location=[22.302711,114.177216], tiles='cartodbpositron', zoom_start=11)
# m_1
```

```python
trees_df = pd.read_csv('/kaggle/input/hong-kong-old-and-stonewall-tree-distribution/ovt_en.csv')
```

```python
import geopandas as gpd

trees = gpd.GeoDataFrame(trees_df, geometry=gpd.points_from_xy(trees_df.XCoordinate, trees_df.YCoordinate))
# print(trees.head())
```

```python
#print(trees.crs)
trees.crs = {'init': 'epsg:2326'}
#print(trees.crs)
trees = trees.to_crs(epsg=4326)
#print(trees.crs)
# print(trees.head())
```

```python
mc = MarkerCluster()
for idx, row in trees.iterrows():
    mc.add_child(Marker([row['geometry'].y, row['geometry'].x]))

m_1.add_child(mc)

# Display the map


m_1
```

```python
trees['lat'] = trees['geometry'].y
trees['long'] = trees['geometry'].x

from folium.plugins import HeatMap
m_2 = folium.Map(location=[22.302711,114.177216], tiles='cartodbpositron', zoom_start=11)

HeatMap(data=trees[['lat', 'long']], radius=10).add_to(m_2)

# Display the map
m_2
```
